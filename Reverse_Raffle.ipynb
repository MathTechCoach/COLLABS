This is a reverse raffle. Enter the number of tickets in the code line below.

```python
numberTickets = 100
```

This script will keep pulling numbers out of the list of numbers 1 through 10 until just one number is left.


```python
import random

ticketList = list(range(1,numberTickets+1))

print(ticketList)
while len(ticketList)>1:
  numPick = random.randint(1,numberTickets)
  if numPick in ticketList:
    print("Remove ",numPick)
    ticketList.remove(numPick)
    print("Numbers left ",ticketList)



```
